```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canadian_immegration_data.csv')
```

```python
df.set_index('Country', inplace=True)
```

```python
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brazil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
brasil_df = pd.DataFrame(brasil_dict)
```

```python
plt.figure(figsize=(8,4))
plt.plot(brasil_df['ano'], brasil_df['imigrantes'])
plt.title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([0,500,1000,1500,2000,2500,3000])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(brasil_df['ano'], brasil_df['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```
